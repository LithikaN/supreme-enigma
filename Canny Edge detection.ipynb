```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img_path = '/content/MainAfter.jpg'
img = cv2.imread(img_path)
if img is None:
    raise FileNotFoundError(f"The file '{img_path}' was not found.")

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define the threshold range for Canny edge detection
lower_threshold = 50  # You can customize this
upper_threshold = 500  # You can customize this

# Apply Canny edge detection
edges = cv2.Canny(gray_img, lower_threshold, upper_threshold, apertureSize=3, L2gradient=True)

# Create an overlay by combining edges with the original image
overlay = cv2.addWeighted(img, 0.8, cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), 0.2, 0)

# Save the output images
output_edge_path = '/content/MainAfter.png'
output_overlay_path = '/content/MainAfter.png'

cv2.imwrite(output_edge_path, edges)
cv2.imwrite(output_overlay_path, overlay)

# Plot the original, grayscale, edge-detected, and overlay images side-by-side
plt.figure(figsize=(15, 10))

# Original Image
plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Grayscale Image
plt.subplot(2, 2, 2)
plt.title('Grayscale Image')
plt.imshow(gray_img, cmap='gray')
plt.axis('off')

# Edge-detected Image
plt.subplot(2, 2, 3)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')
plt.axis('off')

# Overlay Image
plt.subplot(2, 2, 4)
plt.title('Overlay of Edges on Original Image')
plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show the results
plt.tight_layout()
plt.show()

# Print save locations
print(f"Edge-detected image saved at: {output_edge_path}")
print(f"Overlay image saved at: {output_overlay_path}")
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Load the image
img_path = '/content/premium_photo-1664474619075-644dd191935f.jpg'
img = cv2.imread(img_path)
if img is None:
    raise FileNotFoundError(f"The file '{img_path}' was not found.")

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define thresholds for Canny edge detection
lower_threshold = 100
upper_threshold = 300

# Apply Canny edge detection
edges = cv2.Canny(gray_img, lower_threshold, upper_threshold, apertureSize=3, L2gradient=True)

# Create a bold version of the edges by dilating them
kernel = np.ones((3, 3), np.uint8)
bold_edges = cv2.dilate(edges, kernel, iterations=1)

# Create a custom colormap for edges
cmap = ListedColormap(['black', 'cyan', 'magenta', 'yellow'])

# Create a masked version of the original image with edges
masked_img = cv2.bitwise_and(img, img, mask=edges)

# Combine all images into a single figure for visualization
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

axs[0, 1].imshow(gray_img, cmap='gray')
axs[0, 1].set_title('Grayscale Image')
axs[0, 1].axis('off')

axs[1, 0].imshow(bold_edges, cmap=cmap)
axs[1, 0].set_title('Bold Canny Edges (Color-Coded)')
axs[1, 0].axis('off')

axs[1, 1].imshow(cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title('Edges Masked on Original Image')
axs[1, 1].axis('off')

plt.tight_layout()

# Save the results
output_dir = '/content/'
summary_path = f"{output_dir}summary_image_{lower_threshold}_{upper_threshold}.png"
edge_path = f"{output_dir}edges_{lower_threshold}_{upper_threshold}.png"
masked_path = f"{output_dir}masked_image_{lower_threshold}_{upper_threshold}.png"

cv2.imwrite(edge_path, bold_edges)
cv2.imwrite(masked_path, masked_img)
plt.savefig(summary_path)

# Display the final visualization
plt.show()

# Output the file paths
print(f"Summary image saved at: {summary_path}")
print(f"Edge-detected image saved at: {edge_path}")
print(f"Masked image saved at: {masked_path}")
```
